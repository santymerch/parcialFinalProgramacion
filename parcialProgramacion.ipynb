```python
import ipywidgets as widgets
from IPython.display import display

conglomerado = widgets.Text(
    description="Conglomerado:",
    placeholder="Ej: C-12"
)

brigada = widgets.Text(
    description="Brigada:",
    placeholder="Ej: Brigada 1"
)

arboles = widgets.IntText(
    description="Árboles:",
    value=0
)

aves = widgets.IntText(
    description="Aves:",
    value=0
)

area = widgets.FloatText(
    description="Área (ha):",
    value=0.0
)

display(conglomerado, brigada, arboles, aves, area)
```
