```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import json
import os

ARCHIVO = "inspecciones.txt"

def cargar_datos():
    if os.path.exists(ARCHIVO):
        with open(ARCHIVO, "r") as f:
            return json.load(f)
    return []

def guardar_datos(datos):
    with open(ARCHIVO, "w") as f:
        json.dump(datos, f, indent=4)

conglomerado = widgets.Text(
    description="Conglomerado:"
)

brigada = widgets.Text(
    description="Brigada:"
)

arboles = widgets.IntText(
    description="Árboles:"
)

aves = widgets.IntText(
    description="Aves:"
)

area = widgets.FloatText(
    description="Área (ha):"
)

btn_guardar = widgets.Button(description="Guardar")
btn_editar = widgets.Button(description="Editar")
btn_eliminar = widgets.Button(description="Eliminar")
btn_mostrar = widgets.Button(description="Mostrar")

output = widgets.Output()

def guardar(b):
    datos = cargar_datos()

    nueva_inspeccion = {
        "conglomerado": conglomerado.value,
        "brigada": brigada.value,
        "arboles": arboles.value,
        "aves": aves.value,
        "area": area.value
    }

    datos.append(nueva_inspeccion)
    guardar_datos(datos)

    with output:
        clear_output()
        print("Inspección guardada")

btn_guardar.on_click(guardar)

def editar(b):
    datos = cargar_datos()

    encontrado = False

    for inspeccion in datos:
        if inspeccion["conglomerado"] == conglomerado.value:

            inspeccion["brigada"] = brigada.value
            inspeccion["arboles"] = arboles.value
            inspeccion["aves"] = aves.value
            inspeccion["area"] = area.value

            encontrado = True
            break

    guardar_datos(datos)

    with output:
        clear_output()

        if encontrado:
            print("Inspección editada")
        else:
            print("Conglomerado no encontrado")

btn_editar.on_click(editar)

def eliminar(b):
    datos = cargar_datos()

    nuevos_datos = [
        i for i in datos
        if i["conglomerado"] != conglomerado.value
    ]

    guardar_datos(nuevos_datos)

    with output:
        clear_output()
        print("Inspección eliminada")

btn_eliminar.on_click(eliminar)

def mostrar(b):
    datos = cargar_datos()

    with output:
        clear_output()

        if len(datos) == 0:
            print("No hay inspecciones registradas")
        else:
            for i, inspeccion in enumerate(datos, 1):
                print(f"Inspección {i}")
                print(f"Conglomerado: {inspeccion['conglomerado']}")
                print(f"Brigada: {inspeccion['brigada']}")
                print(f"Árboles: {inspeccion['arboles']}")
                print(f"Aves: {inspeccion['aves']}")
                print(f"Área: {inspeccion['area']} ha")
                print("-" * 30)

btn_mostrar.on_click(mostrar)

display(
    conglomerado,
    brigada,
    arboles,
    aves,
    area,
    btn_guardar,
    btn_editar,
    btn_eliminar,
    btn_mostrar,
    output
)
```
